```python
# set return specific gu data
#
# - Parameter: (data_frame: DataFrame ,column_name: "string", law_code : String)
# - Returns: gu data
def get_gu_data(dataframe,column_name, law_code):
    dataframe = dataframe.loc[:][dataframe[column_name] == law_code]
    
    
    mask = dataframe['deal_day'].str.contains('1~10')
    mask2 = dataframe['deal_day'].str.contains('11~20')
    mask3 = dataframe['deal_day'].str.contains('21~')
    column_name = 'deal_day'
    
    dataframe.loc[mask, column_name] = 1
    dataframe.loc[mask2, column_name] = 11
    dataframe.loc[mask3, column_name] = 21
        
    return dataframe
```

```python
# set make full_deal_date & full_addr
#
# - Parameter: (data_frame: DataFrame)
# - Returns: gu data

def make_full_date(dataframe):
    dataframe.loc[:,'full_deal_date'] = dataframe['deal_year'] +"/" + dataframe['deal_month'].astype(str) +"/"+ dataframe['deal_day'].astype(str)
    
    dataframe.loc[:,'full_addr'] = '서울특별시 성북구 ' + dataframe['law_code'] +" "+ dataframe['ji_bun']
    
    dataframe.to_csv('addr_sungbuk.csv', na_rep=0, index=False)
    
    return dataframe
    
    
    
```
